```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import random
import os

from sklearn.cluster import KMeans
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Fare'].fillna(df['Fare'].mean(), inplace=True)
```

```python
df[df['Embarked'].isna()==True]
```

```python
df = df.drop([61,829])
```

```python
df.info()
```

```python
passenger_df = df.drop(['PassengerId', 'Pclass', 'Name', 'Cabin', 'Ticket'], axis=1)
```

```python
passenger_df = pd.get_dummies(passenger_df, columns=['Sex', 'Embarked'])
```

```python
passenger_df.info()
```

```python
passenger_df.head()
```

```python
X = passenger_df
```

```python
sns.countplot(x = "Pclass",hue="Survived", data=df).set_title("Survived by Pclass");
print ("Многие пассажиры не выбрались с нижних палуб.")
```

```python
sns.countplot(x = "SibSp",hue="Survived", data=df).set_title("Survived by Siblings and Spouse");
print ("Много одиноких выжило")
```

```python
sns.set(rc={'figure.figsize':(20,30)})
sns.countplot(y = "Age",hue="Survived", data=df).set_title("Survived by Age");
print ("Выжило много младенцев. Пассажиры от 16 до 50 имеют меньшее значение для выживаемости. Пожилых так же мало выжило")
```

```python
inert = []
for i in range(1,20):
    inert.append(KMeans(n_clusters=i, random_state=2718).fit(X).inertia_)
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.lineplot(range(1,20), inert);
```

```python
km = KMeans(n_clusters=5, random_state=2718)
clusters = km.fit_predict(X)
```

```python
df['clusters'] = clusters
```

```python
sns.barplot(x=clusters, y=df.Fare).set_title('Ticket Price');
```

```python
sns.countplot(data=df, x='clusters', hue='Embarked').set_title('Embarked Code');
```

```python
sns.countplot(data=df, x='clusters', hue='Pclass').set_title('Ticket Class');
```

```python
sns.barplot(x=clusters, y=df.SibSp).set_title('Number of Siblings/Close Family');
```

```python
sns.barplot(x=clusters, y=df.Parch).set_title('Number of Parents/Children');
```

```python
sns.countplot(data=df, x='clusters', hue='Sex').set_title('Sex');
```

```python
sns.barplot(x=df.clusters, y=df.Survived).set_title('Survived');
```
